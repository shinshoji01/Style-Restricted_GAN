```python
fe_list = ["inception", "facial"]#, "facial-feature_reduction"]
metrics = ["precision", "recall", "density", "coverage"]
classes = tuple(range(4))
```

```python
path_list = [
#     "./data/results/62/GAN_evalation_ConventionalSingleGAN.pkl",
#     "./data/results/71/GAN_evalation_DiversifiedSingleGAN.pkl",
#     "./data/results/75/GAN_evalation_ControlGAN_pretrained.pkl",
#     "./data/results/75/GAN_evalation_ControlGAN_relu.pkl",
#     "./data/results/75/GAN_evalation_ControlGAN_0.1.pkl",
#     "./data/results/75/GAN_evalation_ControlGAN_10.pkl",
#     "./data/results/75/GAN_evalation_ControlGAN_1000.pkl",
#     "./data/results/75/GAN_evalation_ControlGAN_conventionalKL.pkl"
    
    "./data/results/62/GAN_evalation_ConventionalSingleGAN-class.pkl",
    "./data/results/75/GAN_evalation_ControlGAN-class_pretrained.pkl",
    "./data/results/75/GAN_evalation_ControlGAN-class_relu.pkl",
]

model_name = []
for path in path_list:
#     model_name.append(os.path.basename(path).split(".")[0][14:])
    model_name.append(os.path.basename(path)[14:-4])
```

```python
source_label = 3
target_label = 2
model_idx = 0

print(model_name[model_idx])
GAN_eval = pickle_load(path_list[model_idx])
scores = []
for fe in fe_list:
    for metric in metrics:
        scores.append(np.mean(GAN_eval[fe][source_label][target_label][metric]))
pd.DataFrame(np.array(scores).reshape(2,4), index=fe_list, columns=metrics)
```
